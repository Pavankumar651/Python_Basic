```python
# Calculate bill: first 100 units free, next 100 @5/unit, above @10/unit
units = int(input("Enter units consumed: "))
# Your code here
if units <= 100:
    bill = 0

elif units <= 200:
    bill = (units - 100) * 5

else:
    bill = (100 * 5) + (units - 200) * 10

print(f"Your bill is ₹{bill}")
```

```python
# Check if withdrawal is possible (balance check, minimum balance 500)
balance = 5000
withdrawal = int(input("Enter amount to withdraw: "))
# Your code here
if withdrawal > balance:
    print("Insufficient balance")

elif balance - withdrawal < 500:
    print("Withdrawal not allowed! Minimum balance of 500 must be maintained")

else:
    balance -= withdrawal
    print(f"Withdrawal successful. Remaining balance: {balance}")
```

```python
# Problem 14: Discount Calculator
# Apply discount: >5000 20%, >2000 10%, else no discount
amount = float(input("Enter purchase amount: "))
# Your code here
if amount > 5000:
    discount = amount * 0.20
elif amount > 2000:
    discount = amount * 0.10
else:
    discount = 0

final_amount = amount - discount

print(f"Original Amount: ₹{amount}")
print(f"Discount: ₹{discount}")
print(f"Final Amount to Pay: ₹{final_amount}")
```

```python
# Calculate BMI and category (Underweight, Normal, Overweight, Obese)

weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))


bmi = weight / (height ** 2)

print(f"Your BMI is: {bmi:.2f}")


if bmi < 18.5:
    print("Category: Underweight")
elif bmi < 25:
    print("Category: Normal")
elif bmi < 30:
    print("Category: Overweight")
else:
    print("Category: Obese")
```

```python
25 # Reverse a given number using while loop

num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10         
    reverse = reverse * 10 + digit
    num = num // 10   

print("Reversed number:", reverse)
```

```python
# Problem 14: Character Frequency Task: Count how many times each character appears in a string.
# Your code here # Example: "hello" → h:1, e:1, l:2, o:1
text = input("Enter text: ")
freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

for ch in freq:
    print(ch, ":", freq[ch])
```
